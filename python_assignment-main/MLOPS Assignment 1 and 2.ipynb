```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class TitanicEDA:
    def __init__(self, file_path):
        self.data = pd.read_csv(file_path)
    
    def summary_statistics(self):
        return self.data.describe(include='all')
    
    def visualize_survival_by_feature(self, feature, save_path=None):
        plt.figure(figsize=(10, 6))
        sns.countplot(data=self.data, x=feature, hue='Survived', palette='viridis')
        plt.title(f'Survival Distribution by {feature}')
        plt.xlabel(feature)
        plt.ylabel('Count')
        
        if save_path:
            plt.savefig(save_path)
        
        plt.show()
    
    def visualize_age_distribution(self, save_path=None):
        plt.figure(figsize=(10, 6))
        sns.histplot(data=self.data, x='Age', hue='Survived', multiple='stack', kde=False, palette='viridis', bins=30)
        plt.title('Age Distribution by Survival')
        plt.xlabel('Age')
        plt.ylabel('Count')
        
        if save_path:
            plt.savefig(save_path)
        
        plt.show()


if __name__ == "__main__":
    file_path = 'titanic.csv'  
    eda = TitanicEDA(file_path)

    # Get summary statistics
    print(eda.summary_statistics())

    # Visualize survival by 'Pclass'
    eda.visualize_survival_by_feature('Pclass', save_path='pclass_survival.png')

    # Visualize survival by 'Sex'
    eda.visualize_survival_by_feature('Sex', save_path='sex_survival.png')

    # Visualize age distribution
    eda.visualize_age_distribution(save_path='age_distribution.png')
```
